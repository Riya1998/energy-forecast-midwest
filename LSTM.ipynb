```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
train = df.iloc[:2000,2]#.to_numpy()
test = df.iloc[2000:,2]#.to_numpy()
print(len(train))
print(len(test))
```

```python
# modifying the shape of the data for lstm

def create_seq(seq, n):
    x = []
    y = []
    for i in range(len(seq)):
        
        k = i + n
        if k > len(seq)-1:
            break
        
        x1 = seq.iloc[i:k]#, -2]
        y1 = seq.iloc[k]#, 2]
        x.append(x1)
        y.append(y1)
        
    return array(x), array(y)
```

```python
x, y = create_seq(df, n)
x = x.reshape((x.shape[0], x.shape[1], 1))
'''for i in range(len(x)):
    print(x[i], y[i])'''
```

```python
n = 12
model = Sequential()
model.add(LSTM(30, activation='relu', input_shape=(n, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
from numpy import array

train_x, train_y = create_seq(train, n)
test_x, test_y = create_seq(test, n)
train_x = train_x.reshape((train_x.shape[0], train_x.shape[1], 1))
test_x = test_x.reshape((test_x.shape[0], test_x.shape[1], 1))

train_x = np.asarray(train_x).astype('float32')
test_x = np.asarray(test_x).astype('float32')

model.fit(train_x, train_y, epochs=200, batch_size = 32, verbose=0)
```

```python
yhat = model.predict(test_x, verbose=0)
print(yhat)
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
mean_absolute_percentage_error(test_y, yhat)
```
